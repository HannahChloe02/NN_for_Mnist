```python
from sklearn.metrics import confusion_matrix  
import numpy as np  
  
# 假设你有以下真实标签和预测标签  
y_true = [0, 1, 1, 1, 0, 1, 2, 1, 0, 1]  # 真实标签，0和1代表两个不同的类别  
y_pred = [0, 0, 2, 1, 2, 1, 1, 1, 0, 1]  # 模型预测的标签  
  
# 使用confusion_matrix函数计算混淆矩阵  
cm = confusion_matrix(y_true, y_pred)  
  
# 输出混淆矩阵  
print("Confusion Matrix:")  
print(cm)
```

```python
import torch  
  
# 假设你有以下tensor形式的真实标签和预测标签  
y_true = torch.tensor([0, 1, 2, 0, 1, 2, 0, 1, 2])  # 真实标签的tensor  
y_pred = torch.tensor([0, 2, 1, 0, 0, 1, 2, 2, 2])  # 模型预测的标签的tensor  
  
# 在PyTorch中，你可以使用自定义函数来生成混淆矩阵  
def confusion_matrix(y_true, y_pred, num_classes):  
    # 初始化混淆矩阵  
    cm = torch.zeros(num_classes, num_classes)  
      
    # 遍历标签  
    for t, p in zip(y_true, y_pred):  
        # 增加对应位置的计数  
        cm[t, p] += 1  
      
    return cm  
  
# 获取类别数量  
num_classes = int(torch.max(y_true).item() + 1)  
  
# 生成混淆矩阵  
cm = confusion_matrix(y_true, y_pred, num_classes)  
  
# 将tensor转换为NumPy数组以便打印和查看  
cm_numpy = cm.numpy()  
  
# 输出混淆矩阵  
print("Confusion Matrix:")  
print(cm_numpy)
```
